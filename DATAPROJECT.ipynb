```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
import geopandas as gpd
import shapely
import re
from datetime import datetime
```

```python
%matplotlib inline
```

```python
plt.rcParams["figure.figsize"] = (10, 5)
```

```python
saved_style_state = matplotlib.rcParams.copy() 
```

```python
if os.path.isfile("ocean_data.xls"):
    filepath = "ocean_data.xls"
    print("loading from file")
else:
    filepath = "https://samoa-data.sprep.org/resource/strandings-oceania-database-data-export"
    print("loading from the internet")

whale_stranding_in_oceania = pd.read_excel(filepath)
print("done")
```

```python
whale_stranding_in_oceania.species.value_counts()
```

```python


pdvc = whale_stranding_in_oceania.species.value_counts() 
pdvc.plot(kind="bar")

plt.title("number of sea-mammal stranded between 1991 - 2021 in oceania", fontsize=15)
plt.xlabel("stranded sea-mammals", fontsize=26)
plt.ylabel("number", fontsize=16)
plt.grid(True)
plt.show()
```

```python


pdvc = whale_stranding_in_oceania.lifeStage.value_counts() 
pdvc.plot(kind="pie")


plt.title(" precentage of marinemammal stranded between 1991 - 2021 in oceania", fontsize=15)
plt.ylabel("lifestage", fontsize=16)
plt.grid(True)
plt.show()
```

```python
pdvc = whale_stranding_in_oceania.sex.value_counts() 
pdvc.plot(kind="pie")

plt.title("precentage of sea-mammal stranded between 1991 - 2021 in oceania ", fontsize=15)
plt.ylabel("gender", fontsize=16)
plt.grid(True)
plt.show()
```

```python
whale_stranding_in_oceania.country.value_counts()
```

```python
pdvc = whale_stranding_in_oceania.country.value_counts() 
pdvc.plot(kind="pie")

plt.title("country of sea-mammal stranded between 1991 - 2021 in oceania ", fontsize=15)
plt.ylabel("country", fontsize=16)
plt.grid(True)
plt.show()
```

```python

number_of_bins = 77
lower_bound = 1
upper_bound = 12

plt.hist(whale_stranding_in_oceania.month, bins=number_of_bins, range=(lower_bound, upper_bound))
plt.title("number of stranded sea-mammal during 1991 - 2021 in oceania according to month", fontsize=15)
plt.xlabel("month", fontsize=6)
plt.ylabel("cases", fontsize=16)
plt.grid(True)
plt.show()


```
